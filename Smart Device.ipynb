```python
class SmartDevice:
    def __init__(self, device_id,power_state,name):
        self.__device_id = device_id
        self.__power_state = power_state
        self.name = name

    @ property
    def device_id(self):
        return self.__device_id

    @ property
    def power_state(self):
        return self.__power_state

    @ device_id.setter
    def device_id(self,device_id):
        if device_id == device_id:
            print(f"Device id is {device_id}")
        else:
            print("Device id is invalid")

    @ power_state.setter
    def power_state(self,power_state):
        if power_state == power_state:
            print(f"Power state is {power_state}")
        else:
            print("Power state is invalid")

    def turn_on(self):
        self.__power_state = "ON"
        print(f"{self.__power_state} is turned on")

    def turn_off(self):
        self.__power_state = "OFF"
        print(f"{self.__power_state} is turned off")

    def display_info(self):
        print(f"Device ID:{self.__device_id}")
        print(f"Power State:{self.__power_state}")
        print(f"Name:{self.name}")

class TemperatureSensor(SmartDevice):
    def __init__(self,device_id, power_state,name, temperature):
        super().__init__(device_id,power_state,name)
        self.temperature = temperature

    def read_temperature(self):
        print(f"Temperature is {self.temperature}")

    def display_info(self):
        super().display_info()
        print(f"Temperature:{self.temperature}")

class SecurityCamera(SmartDevice):
    def __init__(self,device_id,power_state,name,recording_status):
        super().__init__(device_id, power_state, name)
        self.recording_status = recording_status

    def start_recording(self):
        self.recording_status = "Recording"
        print(f"{self.recording_status} has started")

    def stop_recording(self):
        self.recording_status = "Stopped"
        print(f"{self.recording_status} has stopped")

    def display_info(self):
        super().display_info()
        print(f"Recording Status:{self.recording_status}")

class SmartLight(SmartDevice):
    def __init__(self,device_id,power_state,name,brightness):
        super().__init__(device_id,power_state,name)
        self.brightness = brightness

    def display_info(self):
        super().display_info()
        print(f"Brightness:{self.brightness}%")

sensor = TemperatureSensor("AB123","OFF","Living Room Temperature Sensor",25)
sensor.display_info()
light = SmartLight("BC345","OFF","Kitchen Light", 0)
light.display_info()
camera = SecurityCamera("DC678","OFF","Front Door Camera","Stopped")
camera.display_info()

#Menu-Driven Interface
while True:
    print("\n=====SMART DEVICE MANAGEMENT SYSTEM=====")
    print("1.Display Device Information")
    print("2.Turn Device On")
    print("3.Turn Device Off")
    print("4.Read Temperature")
    print("5.Adjust Brightness")
    print("6.Start Recording")
    print("7.Exit")

    choice = int(input("Enter your choice:"))
    if choice == 1:
        print("\nTemperature Sensor")
        sensor.display_info()
        print("\nSmart Light")
        light.display_info()
        print("\nSecurity Camera")
        camera.display_info()
    elif choice == 2:
        device = input("Enter sensor/light/camera: ").lower()
        if device == "sensor":
            sensor.turn_on()
        elif device == "light":
            light.turn_on()
        elif device == "camera":
            camera.turn_on()
        else:
            print("Invalid Input")
    elif choice == 3:
        device = input("Enter sensor/light/camera: ").lower()
        if device == "sensor":
            sensor.turn_off()
        elif device == "light":
            light.turn_off()
        elif device == "camera":
            camera.turn_off()
        else:
            print("Invalid Input")
    elif choice == 4:
        sensor.read_temperature()
    elif choice == 5:
        brightness = input("Enter brightness(0-100): ").lower()
        light.brightness = brightness
        print("Brightness Updated")
    elif choice == 6:
        camera.start_recording()
    elif choice == 7:
        print("Exit")
        break
    else:
        print("Invalid Input")
```
